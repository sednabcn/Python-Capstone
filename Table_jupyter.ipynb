```python
import urllib.request
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
post_list_html=urllib.request.urlopen(url).read()
```

```python
post_data=pd.read_html(post_list_html)
len(post_data)
```

```python
dp1=post_data[0]
dp1.columns
dp1.index
```

```python
dp1.head()
```

```python
dp1.tail()
```

```python
dp=dp1[dp1["Borough"]!="Not assigned"]
```

```python
print(dp.head(10))
```

```python
dpc=dp.copy()
```

```python
dp.Neighbourhood[dpc.Neighbourhood=="Not assigned"]=dpc.loc[dpc.Neighbourhood=="Not assigned",'Borough']
```

```python
dp.head(10)
```

```python
mask=dp.Postcode.duplicated(keep=False)
```

```python
dp[mask]
```

```python
dp_singles=dp[~mask]
```

```python
dp_singles.head(10)
```

```python
dp_duplicates=dp[mask].drop_duplicates(subset="Postcode",keep="first")
```

```python
dp_duplicates
```

```python
f=lambda x: ','.join(z for z in dp.loc[dp.Postcode==x,'Neighbourhood']) 
```

```python
for x in dp_duplicates.index:
    dp_duplicates.Neighbourhood[x]=f(dp_duplicates.Postcode[x])
```

```python
dp_duplicates
```

```python
dp_full = pd.concat([dp_singles,dp_duplicates], ignore_index='True')
```
